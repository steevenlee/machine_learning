From Andrew NG's Machine Learning class, Normal Equation is the fastest approach if number of parameter not too much.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x * (1 +  np.sin(x) / np.pi)
    # return  x * 1.0
    # return  x * (1.2 - 0.4 * np.random.random(x.shape))
    # return  x + N * (0.2 - 0.4 * np.random.random(x.shape))

class Regression:
    def __init__(self,train_x, train_y):
        self.train_x = np.column_stack((np.ones(train_x.shape), train_x))
        self.train_y = train_y

    def linearRegression(self):
        xTx = self.train_x.T.dot(self.train_x)
        print xTx
        if np.linalg.det(xTx) ==0.0:
            print("singular")
            return
        self.w = np.linalg.inv(xTx).dot(self.train_x.T).dot(self.train_y)
        print "w: ", self.w
        
    def calc_test(self, test_x):
        test_x = np.column_stack((np.ones(test_x.shape), test_x))
        pre_y = test_x.dot(self.w)
        return pre_y

    def bias_variance(self, pre_y):
        var = np.mean((pre_y - np.mean(pre_y)) ** 2)
        bias2 = np.mean(np.mean(pre_y) - self.train_y) ** 2
        noise = np.mean((pre_y - self.train_y) ** 2)
        print("var = ", var)
        print("bias2 = ", bias2)
        print("noise = ", noise)
        
    def plotAns(self,pre_y):
        fig = plt.figure(figsize=(15, 8))
        plt.plot(self.train_y, 'b--', label='Training data')
        # plt.plot(self.test_y, 'r-', label='Original data')
        plt.plot(pre_y, label='Fitted line')
        plt.legend()
        plt.show()

N = 50

if __name__=="__main__":
    x = np.arange(N)
    r = Regression(x, f(x))
    r.linearRegression()
    pre_y = r.calc_test(np.arange(N*1.5))
    # r.bias_variance(pre_y)
    r.plotAns(pre_y)
    # print(r.train_x)
    # print(r.train_y)

```

```python
## import numpy as np
import matplotlib.pyplot as plt


N=30

def f(x):
    # return x+x*2*np.sin(x)
    # return x * (1.2 - 0.4 * np.random.random(x.shape))
    return x + N * (0.2 - 0.4 * np.random.random(x.shape))
    # return x * 1.

class regression:
    def __init__(self,train_x,train_y,test_x,test_y):
        self.test_x = test_x
        self.train_x = train_x
        self.train_y = train_y
        self.test_y = test_y

    def linearRegression(self):
        xTx = train_x.T.dot(train_x)
        print "xT * x: ", xTx
        if np.linalg.det(xTx) ==0.0:
            print("singular")
            return
        w = np.linalg.inv(xTx).dot(self.train_x.T).dot(self.train_y)
        print "w: ", w
        pre_y = self.test_x.dot(w)
        return pre_y

    def bias_variance(self,pre_y):
        var = np.mean((pre_y - np.mean(pre_y)) ** 2)
        bias2 = np.mean(np.mean(pre_y) - self.test_y) ** 2
        noise = np.mean((pre_y - self.test_y) ** 2)
        print("var = ", var)
        print("bias2 = ", bias2)
        print("noise = ", noise)
    def plotAns(self,pre_y):
        fig = plt.figure(figsize=(15, 8))
        plt.plot(self.train_y, 'b--', label='Training data')
        plt.plot(self.test_y, 'r--', label='Original data')
        plt.plot(pre_y, label='Fitted line')
        plt.legend()
        plt.show()

if __name__=="__main__":
    x = np.arange(N)
    train_y = f(x)
    # test_x = x + np.random.random()
    train_x = np.column_stack((x, np.ones(N)))

    test_x = np.column_stack((x, np.ones(N)))
    test_y = f(x)


    regressor = regression(train_x,train_y,test_x,test_y)
    pre_y = regressor.linearRegression()
    regressor.bias_variance(pre_y)
    regressor.plotAns(pre_y)
```
